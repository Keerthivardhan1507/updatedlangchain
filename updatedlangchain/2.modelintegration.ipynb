```python
from langchain_groq import ChatGroq

model = ChatGroq(model="qwen/qwen3-32b")
response = model.invoke("what is the number 1 test team in 2017?")
response
```

STREAMING

```python
model.invoke("Write a 200 words paragraph on Artificial Intelligence ?")
```

```python
for chunk in model.stream("Write a 200 words paragraph on Artificial Intelligence ?"):
    print(chunk.text , end= "|",flush = True)
```

```python
for chunk in model.stream("Why do parots have colorful Feathers"):
    print(chunk.text,end = " ")
```

BATCH

```python
responses = model.batch(
    [
        "Who has the highest goals in FIFA world cup?",
        "Why are not india part of football worldcup?",
        "which Game is most popular in India?"
    ]
)
for response in responses:
    print(response)
```

```python
responses = model.batch(
    [
        "Who has the highest goals in FIFA world cup?",
        "Why are not india part of football worldcup?",
        "which Game is most popular in India?"
    ],
    config = {
        'max_concurrency':5
    }
)
for response in responses:
    print(response)
```
