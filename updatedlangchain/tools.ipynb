```python
import os
from dotenv import load_dotenv

load_dotenv()
from langchain.chat_models import init_chat_model

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
model = init_chat_model("groq:qwen/qwen3-32b")
response= model.invoke("why parrots talk?")
response
```

```python
from langchain.tools import tool

@tool
def get_weather(location:str)->str:
    '''Get weather at a location'''
    return f"Its sunny in {location}"

model_with_tools= model.bind_tools([get_weather])
    
```

```python
response= model_with_tools.invoke("Whats the weather in the Boston?")
for tool_call in response.tool_calls:
    print(f"Tool:{tool_call['name']}")
    print(f"Args:{tool_call['args']}")
```

TOOLS EXECUTION LOOP

```python
messages= [{"role":"user","content":"Whats the weather in Boston?"}]
ai_msg = model_with_tools.invoke(messages)
messages.append(ai_msg)

for tool_call in ai_msg.tool_calls:
    tool_result = get_weather.invoke(tool_call)
    messages.append(tool_result)
final_response = model_with_tools.invoke(messages)
print(final_response.text)
```
